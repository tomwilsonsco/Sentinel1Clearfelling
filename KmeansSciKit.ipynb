```python
region = 'ScotlandRSRegion3'
wd= '/content/drive/My Drive/colab_data/{}'.format(region)
```

```python
inputImg = os.path.join(wd,'sentine1DiffMask.tif')
```

```python
s1 = rasterio.open(inputImg)

B = s1.count
print(B)
```

```python
inputI = s1.read()
samples = inputI.reshape((4, -1)).T
samples.shape
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=2)
```

```python
labels = clf.fit_predict(samples)
labels.shape
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(labels.reshape((2514, 2299)), cmap="Set1")
```
